## 4. Convertir le dataset expanded en format chat + faire le split (90/10)

```python
import jsonfrom pathlib import Pathfrom sklearn.model_selection import train_test_splitSOURCE = Path('/content/coach_dataset_expanded.jsonl')TRAIN_OUT = Path('/content/coach_dataset_train.jsonl')VAL_OUT = Path('/content/coach_dataset_val.jsonl')SYSTEM_PROMPT = (    "Tu es un coach de prise de parole. "    "Tu dois repondre uniquement en JSON valide. "    "Tu dois toujours utiliser exactement ces 3 cles : bilan_global, point_prioritaire, encouragement. "    "N'utilise jamais d'autres cles. "    "Ne rajoute aucun texte avant ou apres le JSON. "    "Chaque valeur doit etre une phrase ou deux phrases courtes en francais simple, professionnel, toujours avec 'vous'. "    "Tu restes concret, sobre et directement lie aux signaux fournis. "    "Tu n'utilises jamais de formulations lyriques, generiques ou grandiloquentes.")def build_user_content(example):    data = example['input']    top = data['top_recommendation']    strengths = ' | '.join(data.get('strengths', []))    weaknesses = ' | '.join(data.get('weaknesses', []))    return (        f"Score global: {data['overall_score']}\n"        f"Voix: {data['voice_score']}/10\n"        f"Presence: {data['presence_score']}/10\n"        f"Corps: {data['body_language_score']}/10\n"        f"Scene: {data['scene_score']}/10\n"        f"Points forts: {strengths}\n"        f"Axes de progression: {weaknesses}\n"        f"Top recommandation: {top['category']}\n"        f"Message: {top['message']}\n"        f"Conseil actionnable: {top['actionable_tip']}"    )def build_assistant_content(example):    target = example['target']    payload = {        'bilan_global': target['bilan_global'],        'point_prioritaire': target['point_prioritaire'],        'encouragement': target['encouragement'],    }    return json.dumps(payload, ensure_ascii=False)rows = []with SOURCE.open('r', encoding='utf-8') as f:    for line in f:        line = line.strip()        if not line:            continue        example = json.loads(line)        rows.append({            'messages': [                {'role': 'system', 'content': SYSTEM_PROMPT},                {'role': 'user', 'content': build_user_content(example)},                {'role': 'assistant', 'content': build_assistant_content(example)},            ]        })print('Total examples:', len(rows))train_rows, val_rows = train_test_split(rows, test_size=0.1, random_state=42, shuffle=True)with TRAIN_OUT.open('w', encoding='utf-8') as f:    for row in train_rows:        f.write(json.dumps(row, ensure_ascii=False) + '\n')with VAL_OUT.open('w', encoding='utf-8') as f:    for row in val_rows:        f.write(json.dumps(row, ensure_ascii=False) + '\n')print('Train:', len(train_rows))print('Val:', len(val_rows))
```

## 5. Charger les datasets

```python
from datasets import load_dataset

data_files = {
    'train': str(TRAIN_OUT),
    'validation': str(VAL_OUT),
}

dataset = load_dataset('json', data_files=data_files)
dataset
```

## 6. Charger Qwen 2.5 3B avec QLoRA

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

model_name = 'Qwen/Qwen2.5-3B-Instruct'

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map='auto',
    trust_remote_code=True
)

model.config.use_cache = False
```

## 7. Formater le dataset avec le chat template

```python
def format_example(example):
    text = tokenizer.apply_chat_template(
        example['messages'],
        tokenize=False,
        add_generation_prompt=False
    )
    return {'text': text}

dataset = dataset.map(format_example)
dataset['train'][0]['text'][:1000]
```

## 8. Configurer QLoRA

```python
from peft import LoraConfig

peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    bias='none',
    task_type='CAUSAL_LM',
    target_modules=[
        'q_proj',
        'k_proj',
        'v_proj',
        'o_proj',
        'gate_proj',
        'up_proj',
        'down_proj',
    ],
)
```

## 9. Configurer l'entrainement (v2)

```python
from transformers import TrainingArgumentsoutput_dir = '/content/drive/MyDrive/speechcoach_qwen25_3b_qlora_v2'training_args = TrainingArguments(    output_dir=output_dir,    num_train_epochs=3,    per_device_train_batch_size=1,    per_device_eval_batch_size=1,    gradient_accumulation_steps=8,    learning_rate=1e-4,    warmup_steps=5,    logging_steps=5,    eval_steps=10,    save_steps=10,    save_total_limit=2,    evaluation_strategy='steps',    bf16=torch.cuda.is_available(),    fp16=not torch.cuda.is_available(),    report_to='none',    remove_unused_columns=False,)
```

## 10. Lancer l'entrainement

```python
from trl import SFTTrainer

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset['train'],
    eval_dataset=dataset['validation'],
    peft_config=peft_config,
    args=training_args,
    dataset_text_field='text',
    max_seq_length=1024,
)

trainer.train()
```

## 11. Sauvegarder l'adapter et le tokenizer

```python
save_path = '/content/drive/MyDrive/speechcoach_qwen25_3b_qlora_v2/final_adapter'trainer.model.save_pretrained(save_path)tokenizer.save_pretrained(save_path)print('Saved adapter to:', save_path)
```
